```r
wine = read.csv("/home/david/courses/IBM-backend/Python-for-Data-Science-AI-And-Develpment/final-project/wine-quality-ds.csv")
head(wine)
# View(wine) # shows the dataset in a spreadsheet-like viewer, but may not work in all environments

barplot(table(wine$quality), main="Wine Quality Distribution", xlab="Quality", ylab="Frequency")
# View(wine$quality)

# data clasification
wine$taste <- ifelse(wine$quality < 6, 'bab', 'good')
wine$taste[wine$quality == 6] <- 'normal'
wine$taste <- as.factor(wine$taste) # transform the column taste of wine in factor data type

# split the dataset in two one to train and another to test
set.seed(123) # Fija la semilla para que el muestreo sea reproducible y simpre el mismo
samp <- sample(nrow(wine), 0.6 * nrow(wine)) # selecciona aleatoriamente el 60% de los ndices de las filas del dataset wine.
train <- wine[samp, ] # Crea el conjunto de entrenamiento con el 60% de los datos seleccionados.
test <- wine[-samp, ] # Crea el conjunto de prueba con el 40% restante de los datos.

# install the forest data-structure and model
install.packages("randomForest")
library(randomForest)
model <- randomForest(taste ~ . - quality, data=train)

#
model
pred <- predict(model, newdata=test)
table(pred, test$taste)
```
